```python
#Import Pandas and sqlalchemy
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Read in Zipcode CSV
csv_file = "raw_data/us-zip-code-latitude-and-longitude.csv"
zipcode_df = pd.read_csv(csv_file)
zipcode_df.head()
```

```python
#Remove unneeded columns
condensed_df = zipcode_df[["Zip", "State","Latitude","Longitude"]]
condensed_df.head()
```

```python
#Filter dataset by CA and FL
states= ["CA","FL"]
filter_df= condensed_df[condensed_df['State'].isin(states)]
filter_df.head()
```

```python
lat_long_df = filter_df.sort_values(by="Zip", ascending=False)
lat_long_df.head()
```
