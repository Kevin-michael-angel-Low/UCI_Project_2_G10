```python
#Import Pandas and sqlalchemy
import pandas as pd
import numpy as py
from sqlalchemy import create_engine
```

```python
#Read in Zipcode CSV
csv_file = "raw_data/us-zip-code-latitude-and-longitude.csv"
zipcode_df = pd.read_csv(csv_file)
zipcode_df.head()
```

```python
#Remove unneeded columns
condensed_df = zipcode_df[["Zip", "State","Latitude","Longitude"]]
condensed_df.head()
```

```python
#Filter dataset by CA and FL
states= ["CA","FL"]
filter_df= condensed_df[condensed_df['State'].isin(states)]
filter_df.head()
```

```python
#Sort final Zip Code dataset
lat_long_df = filter_df[["Zip","Latitude","Longitude"]]
lat_long_df=lat_long_df.sort_values(by="Zip", ascending=False)
lat_long_df = lat_long_df.reset_index()
lat_long_df.head()
```

```python
#Read in COVID CSV
csv_file2 = "raw_data/COVID_HOSPITAL_DATA_US.csv"
data_df = pd.read_csv(csv_file2)
data_df.head()
```

```python
covid_df= data_df[["collection_week",
                   "state",
                   "zip",
                   "hospital_name",
                   "city",
                   "inpatient_beds_used_7_day_avg",
                   "total_adult_patients_hospitalized_confirmed_covid_7_day_avg",
                  "all_adult_hospital_inpatient_bed_occupied_7_day_avg"]]
covid_df.head()
```

```python
covid_df= covid_df[covid_df['state'].isin(states)]
covid_df.head()
```

```python
covid_df= covid_df.rename(columns={"collection_week":"Week_Data_Collected",
                                   "state": "State",
                                   "zip":"Zip",
                                   "city": "City",
                                   "hospital_name":"Hosptial_Name",
                                   "inpatient_beds_used_7_day_avg":"Total_Beds_Available",
                                   "total_adult_patients_hospitalized_confirmed_covid_7_day_avg":"Total_Covid_Patients",
                                   "all_adult_hospital_inpatient_bed_occupied_7_day_avg":"Total_Beds_in_Use"})
covid_df.head()
                                   
```
